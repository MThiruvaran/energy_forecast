# Importing the libraries



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

# Total Electricity consumption per capita

```python
df = pd.read_csv('SeriesExport-08-05-2023-11-27-49.csv')
df
```

```python
import missingno
missingno.matrix(df)
```

```python
years = np.array(df['YEAR']).reshape(-1, 1)
estpp = np.array(df['ESTPP'])
```

```python
polynomial_degree = 2
```

```python
poly_features = PolynomialFeatures(degree=polynomial_degree)
X_poly = poly_features.fit_transform(years)

model = LinearRegression()
model.fit(X_poly, estpp)

predictions = model.predict(X_poly)
```

```python
plt.scatter(years, estpp, color='blue', label='data')
plt.plot(years, predictions, color='red', label='Polynomial Regression')
plt.xlabel('Year')
plt.ylabel('Electricity per Capita')
plt.title('Polynomial Regression for Electricity per Capita')
plt.legend()
plt.show()
```

```python
r2 = r2_score(estpp, predictions)
print("R-squared:", r2)
```
