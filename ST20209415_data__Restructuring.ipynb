```python

fileNames = ['SeriesExport-08-05-2023-13-14-54.csv']
```

```python
df = pd.read_csv('SeriesExport-08-05-2023-13-14-54.csv')
df
```

```python
for name in fileNames:
  df=pd.read_csv(name)
  rowName = df.columns[1::2]
  rowName = [string[5:10] for string in rowName]
  df = df[6:].reset_index(drop=True)
  colName = list(df['Series Key'])
  rowName.append("YEAR")
  rowValue = df.columns[1::2]
  newTempDf = {
    'YEAR': colName
  }

  for i in range(len(rowName)-1):
    newTempDf[rowName[i]] = list(df[rowValue[i]])

  cleanedDf = pd.DataFrame(newTempDf)
  filePath = "/content/drive/My Drive/cleaned data/{}".format(name)
  cleanedDf.to_csv(filePath, index=False)
```
